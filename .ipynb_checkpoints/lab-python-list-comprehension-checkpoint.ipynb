2. Modify the function get_customer_orders so it prompts the user to enter the number of customer orders and gathers the product names using a loop and user input. Use comprehension.

3. Add a new function to calculate the total price of the customer order. For each product in customer_orders, prompt the user to enter the price of that product. Use comprehension to calculate the total price. Note: assume that the user can only have 1 unit of each product.

Your code should produce output similar to the following:

```python

Order Statistics:
```

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Fonction pour initialiser l'inventaire
def initialize_inventory(products):
    quantities = [int(input(f"Enter the quantity of {product}: ")) for product in products]
    return dict(zip(products, quantities))

# Fonction pour obtenir les commandes du client
def get_customer_orders():
    num_orders = int(input("Enter the number of customer orders: "))
    customer_orders = [input(f"Enter the name of product {i + 1}: ") for i in range(num_orders)]
    return set(customer_orders)

# Fonction pour mettre à jour l'inventaire
def update_inventory(customer_orders, inventory):
    products_to_remove = set()  # Liste des produits à supprimer de l'inventaire
    for product in customer_orders:
        if product in inventory:
            inventory[product] -= 1
            if inventory[product] == 0:
                products_to_remove.add(product)
    
    # Supprime les produits de l'inventaire avec une quantité de zéro
    for product in products_to_remove:
        del inventory[product]

# Fonction pour calculer les statistiques de commande
def calculate_order_statistics(customer_orders, products):
    total_ordered = len(customer_orders)
    unique_products_ordered = len(set(customer_orders))
    return total_ordered, unique_products_ordered

# Nouvelle fonction pour calculer le prix total de la commande du client
def calculate_total_price(customer_orders):
    total_price = sum([float(input(f"Entrez le prix de {product}: ")) for product in customer_orders])
    return total_price

# Fonction pour imprimer les statistiques de commande
def print_order_statistics(order_statistics):
    total_ordered, unique_products_ordered = order_statistics
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {total_ordered}")
    print(f"Percentage of Unique Products Ordered: {unique_products_ordered / len(products) * 100}%")

# Fonction pour imprimer l'inventaire mis à jour
def print_updated_inventory(inventory):
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
        

# Appel des fonctions pour exécuter le programme
inventory = initialize_inventory(products)
customer_orders = get_customer_orders()
update_inventory(customer_orders, inventory)
order_statistics = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_statistics)
print_updated_inventory(inventory)
total_price = calculate_total_price(customer_orders)

print(f"Prix total de la commande du client : {total_price:.2f} €")
````
